```python
from PIL import Image,ImageFilter
import numpy as np
import matplotlib.pyplot as plt
# 读取图像
image = Image.open('./data/original.jpg')

#将图像转换为灰图
image = image.convert('L')
print(np.array(image).shape)

#plt.imshow(image,cmap='gray')
#定义卷积核
kernel = np.array([ 
                    [99999,99,99],
                    [99,99,99],
                    [99,99,999]
                  ])
#做卷积操作
image = image.filter(ImageFilter.Kernel((3, 3), kernel.flatten()) )

#展示卷积后的图像
plt.imshow(image,cmap='gray')
#plt.axis('off')
plt.show()
```
