```python
# # 数据可视化
# # 创建DataFrame,用它来处理展示数据
# data = pd.DataFrame(data=iris.data, columns=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
# data['target'] = iris.target # 添加一个目标列

data = pd.read_excel('./data/中国人寿.xlsx')
data.head()
# data.shape
```

```python
# 概率分布曲线, x=x轴数据; fill=图像是否有阴影；hue=根据什么来划分
data = pd.DataFrame(data)
sns.kdeplot(data,x='charges', fill=True, hue='sex')
```
