```python
import numpy as np
import matplotlib.pyplot as plt
X = np.linspace(-1,11,num=100)
X
```

```python
y = (X - 5)**2 + 3 * X- 12 + np.random.randn(100,1)
# -1:自适应形状 1弄成一列
X = X.reshape(-1,1)
X.shape
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
#使用线性回归预测
model = LinearRegression()
model.fit(X,y) #训练

# 生成测试数据
X_test = np.linspace(-2,12,300).reshape(-1,1)

# 预测
y_pred = model.predict(X_test)
plt.scatter(X,y)

plt.plot(X_test,y_pred, color='green')
```

## 升维(直线可以变成曲线)

```python
np.set_printoptions(suppress=True)
X2 = np.concatenate([X,X**2], axis=1)

# y = (X - 5)**3 + 3 * X- 12 + np.random.randn(100,1)

model2 = LinearRegression()
model2.fit(X2,y)

X_test2 = np.concatenate([X_test,X_test**2], axis=1)
y_pred = model2.predict(X_test2)

plt.scatter(X,y)

plt.plot(X_test,y_pred,color='red')
```
