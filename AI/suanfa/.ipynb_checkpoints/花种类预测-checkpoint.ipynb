```python
from sklearn.datasets import load_iris, fetch_20newsgroups
```

```python
# 数据集获取
iris = load_iris()
print('特征值名字',iris.feature_names)
print("特征值",iris['data'])
print("目标值名字(人打的标签)",iris.target_names)
print("目标值(人打的标签)",iris.target)
print('描述',iris.DESCR)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 数据可视化
# 创建DataFrame,用它来处理展示数据
iris_d = pd.DataFrame(data=iris.data, columns=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'])
iris_d['target'] = iris.target # 添加一个目标列
print(iris_d)

def iris_plot(data, col1, col2):
    sns.lmplot(x=col1, y=col2, data=data, hue='target', fit_reg=False)
    plt.show()
    

# 这个图画出来是啥意思呢????
iris_plot(iris_d, 'sepal length (cm)', 'sepal width (cm)')
```

```python
from sklearn.model_selection import train_test_split
# 数据集划分
#  iris.data:数据集；iris.target数据集目标值；test_size=测试集占比例; random_state=随机？
#  x_trian分割好的训练集, x_test分割好的测试集；y_trian分割好的训练集的目标值(标签),y_test分割好的测试集的目标值(标签)
x_trian, x_test, y_trian, y_test = train_test_split(iris.data,iris.target, test_size=0.2, random_state=22)
print("训练集的目标值形状\n",y_trian.shape)
print("测试集的目标值形状\n",y_test.shape)

print("训练集的特征值\n",x_trian)
print("训练集的标记(标注、标签、目标值)\n",y_trian)
print("测试集的特征值\n",x_test)
print("测试集的目标值\n",y_test)
```
