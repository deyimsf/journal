# 1 生成数组的方法
## 1.1生成0,1数组
  * np.ones(shape,dtype) 生成1数组，shape是数组形状，dtype是数据类型
  * np.ones_like(aaa,dtype) 生成和aaa同形状的1数组 
  
  * np.zeros(shape,dtype) 生成0数组，shape是数组形状，dtype是数据类型
  * np.zeros(aaa,dtype) 生成和aaa同形状的0数组

```python
# 生成1数组
ones = np.ones([4,8]) #4行8列
ones
```

```python
# 生成和ones同形状的数组
ones_copy = np.zeros_like(ones)
ones_copy
```

# 1.2 数组生成方式
* np.array(object,dtype) 深拷贝
* np.asarray(a,dtype) 浅拷贝，感觉就是一个指针引用

```python
a = np.array([[1,2,3],[4,5,6]])
a

a1 = np.array(a)
a2 = np.asarray(a)
a[0,0]=10000
a1,a2
```

## 1.3生成固定范围的数组
### 1.3.1 np.linspace(start,stop,num,endpoint)
* 创建等差数组 -- 按个数
* start 开始值
* end 结束值
* num 要生成的等差数据的数量,默认50
* endpoint 是否包含stop值，默认true

```python
# 0~3之间，生成12个等差数组
np.linspace(0,3,12)
```

## 1.3.2 np.arange(start,stop,step,dtype)
* 生成等差数组 -- 按步长

```python
# 0~3之前，按步长2生成等差数组
np.arange(0,3,2)
```

## 1.3.3 np.logspace(start,stop,num)
* 创建等比数列，10的多少次方的等比数列,其实就是对数
* num 要生成的等比数列数量，默认50

```python
# 在10^0 ~ 10^3之间，生成5个等比数列
np.logspace(0,3,5) #10^0=1,....,10^3=1000;  log1=0,log1000=3
```

# 1.4 生成随机数组
## 1.4.1 np.random.normal(loc,scale,size) 是一个numpy函数，用于获取正态分布的随机数
* loc 正太概率均值； μ
* scale 标准差、均方差 ；σ
* size 要采样的数据个数,或者数组形状？

```python
# 标准正态分布采样
x1 = np.random.normal(0,1,1000000)
x1

# 创建画布
plt.figure(figsize=(20,4), dpi=100)

# 绘制直方图,把x1中的数据分成1000组进行绘制
plt.hist(x1,1000)

# 显示图像
plt.show()
```


## 1.4.2均匀分布

```python
x2 = np.random.uniform(-1,1,1000000)
x2
```

```python
# 创建画布
plt.figure(figsize=(20,4), dpi=100)

# 绘制直方图,把x2中的数据分成1000组进行绘制
plt.hist(x2,1000)

# 显示图像
plt.show()
```

# 3 形状改变
## 3.1 .reshape(shape,order) 返回一个shape形状的数组，原数组不变
* shape 改变形状，数据拍扁后重新排序

```python
stock_changes = np.random.normal(0,1,(4,5))

# 从4,5形状，变成5行4列,但原数组本身不会改变(还是4行5列)
aa = stock_changes.reshape([5,4]),
aa
```

## 3.2 .resize(shape)
* 更改原数组的形状

```python
stock_changes.resize([5,4])
stock_changes
```

## 3.3 .T 数组转置

```python
# 转置前形状
stock_changes.shape
```

```python
# 转置
zhuanzhi = stock_changes.T
zhuanzhi
```

```python
# 转置后的形状
zhuanzhi.shape
```
