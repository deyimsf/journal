# 其它图形绘制(散点图、柱状图等)

```python
import matplotlib.pyplot as plt
import random

from pylab import mpl
# 处理中文不能显示的问题
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 处理有些符号无法正常显示的问题
mpl.rcParams["axes.unicode_minus"] = False
```

# 1 散点图的绘制

```python
# 0准备数据
x = [random.uniform(30,500) for i in range(20)]
y = [random.uniform(30,400) for i in range(20)]

# 1创建画布
plt.figure(figsize=(20,4), dpi=100)

# 2绘制散点图
plt.scatter(x,y)

# 3显示图像并释放资源
plt.show()
```

# 2 柱状图绘制

```python
# 0准备数据
#电影名字
movie_name = ['雷神3','诸神黄昏','电影1','电影2','电影3','电影4','电影5','电影6']
# x轴数据
x = range(len(movie_name))
# 票房数据
y = [3333,44444,55555,6666,77777,44455,66733,67777]

# 1创建画布
plt.figure(figsize=(20,4), dpi=100)

# 2绘制柱状图
plt.bar(x,y,color=['b','r','g','y','c','k','b'], width=0.5)

# 2.1修改x轴显示, 用准备好的刻度数据来刻画坐标轴
plt.xticks(x,movie_name)

# 2.2添加网格
plt.grid(linestyle='--', alpha=0.7)

# 2.3添加标题
plt.title("电影票房收入对比")

# 3图像显示
plt.show()
```
