# 多坐标系实现绘图 (面向对象的画图)

```python
import matplotlib.pyplot as plt
import random

from pylab import mpl
# 处理中文不能显示的问题
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 处理有些符号无法正常显示的问题
mpl.rcParams["axes.unicode_minus"] = False


# 0.准备数据
x = range(60)    #x轴数据 -->  数字从0到59，总共60个数字
y_shanghai = [random.uniform(15,18) for i in x] #y轴数据 --> 随机生成[15,18]之间的实数，总共生成x个(x值是上面的range(60))
y_beijing = [random.uniform(1,3) for i in x]

# 1.创建画布
#                   绘制1行(画布在同一行上)、2列(有列画布)、长20宽4的画布，清晰度是100
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,4), dpi=100)

# 2.绘制图像
axes[0].plot(x, y_shanghai, label="上海")
axes[1].plot(x, y_beijing, color='r', linestyle='--', label="北京") #绘制北京的折线图

# 2.1 准备x,y轴要显示的刻度数据
x_ticks_label = ["11点{}分".format(i) for i in x]  #x轴要显示的刻度数据
y_ticks = range(40)  #y轴要显示的刻度数据，从uniform(15,18)的60个，变成了0~39个

# 2.2 用准备好的刻度数据来刻画坐标轴
axes[0].set_xticks(x[::5]) #修改第一张图x轴刻度
axes[0].set_xticklabels(x_ticks_label[::5]) #第一张图x轴对应刻度复杂显示
axes[0].set_yticks(y_ticks[::5]) #修改第一张图y轴刻度

axes[1].set_xticks(x[::5]) #修改第二张图x轴刻度
axes[1].set_xticklabels(x_ticks_label[::5]) #第二张图x轴对应刻度复杂显示
axes[1].set_yticks(y_ticks[::5]) #修改第二张图y轴刻度


# 2.3 添加网格显示  gird(是否显示网格，线条样式，透明度)
axes[0].grid(True,linestyle="--",alpha=0.5)
axes[1].grid(True,linestyle="--",alpha=0.5)


# 2.3 添加描述信息
axes[0].set_xlabel("时间",fontsize=20) #x轴标签
axes[0].set_ylabel("温度",fontsize=20) #y轴标签
axes[0].set_title("中午11点~12点某城市温度变化图",fontsize=20) #图标标签

axes[1].set_xlabel("时间",fontsize=20) #x轴标签
axes[1].set_ylabel("温度",fontsize=20) #y轴标签
axes[1].set_title("中午11点~12点某城市温度变化图",fontsize=20) #图标标签

# 2.5 显示图例 legend:图例
axes[0].legend(loc=0) #loc=0~10,对应图例显示的位置;也可以是“best”,"upper right","upper left","center"等
axes[1].legend(loc=0) 

# 3.显示图像
plt.show() #画图并释放资源
```
