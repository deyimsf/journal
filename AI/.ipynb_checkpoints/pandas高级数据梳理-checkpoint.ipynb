```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1 缺失值处理

```python
movie = pd.read_csv('./data/IMDB-Movie-Data.csv')
movie.head()
```

```python
# 非空设置为True，null设置为Fasle
pd.notnull(movie)
```

```python
# 都不是null时返回True,有一个是null则返回False
np.all(pd.notnull(movie))
```

```python
# 有空值设置为True，非空为False
pd.isnull(movie)
```

```python
# 删除缺失值
data = movie.dropna()
pd.notnull(data)
```

```python
# 取某列平均值
mean = movie['Revenue (Millions)'].mean()
mean
```

```python
# 把某列中的NaN值替换为平均值
movie['Revenue (Millions)'].fillna(mean, inplace=True) # inplace指定是否在原位置替换
movie.head()
```

```python
# 处理所有列的缺失值
for i in movie.columns:
    if np.any(pd.isnull(movie[i])) == True: #如果这一列有缺失值
        print(i)
        movie[i].fillna(movie[i].mean(), inplace=True)
```

```python
np.any(pd.isnull(movie)) #
```

 # 2 数据离散化

```python
data = pd.read_csv('./data/stock_day.csv')
```

```python
data.head()
```

```python
p_change = data['p_change']
p_change.head()
```

```python
# qcut会将p_change中的数据进行排序，然后再将p_change数据均分成10组,
# 其中每个分组是一个区间，如(1.738, 2.938]是一个区间，下面的2018-02-27、2018-02-23、2018-02-14都在这个区间
qcut = pd.qcut(p_change, 10)
qcut
```

```python
# 看下每个分组中数据个数
qcut.value_counts() #下面数据(1.738, 2.938] 表示在这个区间有64个数据
```

```python
# 自定义分组区间
bins = [-100,-7,-5,-3,0,3,5,6,100]
p_count = pd.cut(p_change,bins) #按照bins中制定的区间进行分组
p_count
```

```python
# 查看每个区间的个数
p_count.value_counts()
```

```python
# one-hot编码，把分类数据弄成one-hot方式
datat = pd.cut(p_change,bins)
pd.get_dummies(datat, prefix="rise").head() # 2018-02-27属于(0, 3]中的数据，所以这个交叉点是True
```

# 4 交叉表透视表

```python
data.head()
```

```python
# 查看行索引数据，目前元素类型是object
data.index
```

```python
# 索引数据类型转换成日期
time = pd.to_datetime(data.index)
time
```

```python
# 查看每个日期对应周几
time.weekday
```

```python
# 查看每个日期对应多好号
time.day
```

```python
# 数据加一列，列名是week
data['week'] = time.weekday
data
```

```python
# 价格大于0(股票涨了)设置为2，小于0(股票跌了)设置为0
p_n = np.where(data['p_change']>0, 2, 0)
p_n
```

```python
# 涨跌数据单独放一列
data['p_n'] = p_n
data.head()
```

```python
# 交叉,统计每周，涨的有多少，跌的有多少
# 先按week进行分组，然后在每个分组内部，再按照p_n字段值进行分组求和
count = pd.crosstab(data['week'], data['p_n'])
count
```

```python
# 按行求和
sum = count.sum(axis=1)
# 转换成浮点数
sum = sum.astype(np.float32)
```

```python
# 展示每个数据所在行的百分比
ret = count.div(sum,axis=0)
ret
```

```python
# 数据画成柱状图  stacked=是否堆积摞起来
ret.plot(kind='bar', stacked=True)
plt.show()
```
