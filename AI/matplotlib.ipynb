# 1.1基础图像功能实现

```python
import matplotlib.pyplot as plt
# random() 方法返回随机生成的一个实数，它在[0,1)范围内。
# 不能直接访问的，需要导入 random 模块，然后通过 random 静态对象调用该方法。
import random 

# range() 创建一个整数列表，一般用在 for 循环中
# 语法 range(start, stop[, step])
#     start:开始数字，默认0
#     stop:结束数字
#     step:步长，默认1

# random.uniform(x,y) 将随机生成下一个实数，它在 [x, y] 范围内。


# 0.准备数据
x = range(60)    #x轴数据 -->  数字从0到59，总共60个数字
y = [random.uniform(15,18) for i in x] #y轴数据 --> 随机生成[15,18]之间的实数，总共生成x个(x值是上面的range(60))

# 1.创建画布 长20宽8 清晰度100
plt.figure(figsize=(20,4),dpi=100)

# 2.绘制图像
plt.plot(x, y)

# 3.显示图像
plt.show()
```

# 1.2实现其他功能（x,y轴刻度修改）

```python
import matplotlib.pyplot as plt
import random

from pylab import mpl
# 处理中文不能显示的问题
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 处理有些符号无法正常显示的问题
mpl.rcParams["axes.unicode_minus"] = False


# 0.准备数据
x = range(60)    #x轴数据 -->  数字从0到59，总共60个数字
y = [random.uniform(15,18) for i in x] #y轴数据 --> 随机生成[15,18]之间的实数，总共生成x个(x值是上面的range(60))
2
# 1.创建画布 长20宽8 清晰度100
plt.figure(figsize=(20,4),dpi=100)

# 2.绘制图像
plt.plot(x, y)

# 2.1 准备x,y轴要显示的刻度数据
x_ticks_lable = ["11点{}分".format(i) for i in x] #x轴要显示的刻度数据
y_ticks = range(40) #y轴要显示的刻度数据，从uniform(15,18)的60个，变成了0~39个

# xticks 获取或设置当前x轴刻度位置和标签，不传参数则显示当前刻度
# 语法: xticks(ticks=[],lables=[],**kwargs)
#      ticks: x轴刻度位置列表;不传则显示当前刻度;传空列表则不显示x轴
#      lables: 和ticks参数值一一对应，一般用来把ticks中对应的数字替换成lables中的字符串
#      **kwargs: 文本属性，用来控制标签文本的展示，例如字体大小和样式       

# 2.2 用准备好的刻度数据来刻画坐标轴
# x[start:end:step],对x列表中取值，start:从哪儿开始; end:到哪儿结束; step:步长
plt.xticks(x[::5],x_ticks_lable[::5]) #修改x轴刻度
plt.yticks(y_ticks[::5]) #修改y轴刻度

# 2.3 添加网格显示  gird(是否显示网格，线条样式，透明度)
plt.grid(True,linestyle="--",alpha=0.5)

# 2.3 添加描述信息
plt.xlabel("时间",fontsize=20) #x轴标签
plt.ylabel("温度",fontsize=20) #y轴标签
plt.title("中午11点~12点某城市温度变化图",fontsize=20) #图标标签

# 2.4 保存图片
# plt.savefig("./test.png")

# 3.显示图像
plt.show() #画图并释放资源
```

# 2 一个坐标系中绘制多个图像

```python
import matplotlib.pyplot as plt
import random

from pylab import mpl
# 处理中文不能显示的问题
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 处理有些符号无法正常显示的问题
mpl.rcParams["axes.unicode_minus"] = False


# 0.准备数据
x = range(60)    #x轴数据 -->  数字从0到59，总共60个数字
y = [random.uniform(15,18) for i in x] #y轴数据 --> 随机生成[15,18]之间的实数，总共生成x个(x值是上面的range(60))
y_beijing = [random.uniform(1,3) for i in x]

# 1.创建画布 长20宽8 清晰度100
plt.figure(figsize=(20,4),dpi=100)

# 2.绘制图像
plt.plot(x, y, label="上海")
plt.plot(x, y_beijing, color='r', linestyle='--', label="北京") #绘制北京的折线图

# 2.1 准备x,y轴要显示的刻度数据
x_ticks_lable = ["11点{}分".format(i) for i in x] #x轴要显示的刻度数据
y_ticks = range(40) #y轴要显示的刻度数据，从uniform(15,18)的60个，变成了0~39个
    
# 2.2 用准备好的刻度数据来刻画坐标轴
plt.xticks(x[::5],x_ticks_lable[::5]) #修改x轴刻度
plt.yticks(y_ticks[::5]) #修改y轴刻度

# 2.3 添加网格显示  gird(是否显示网格，线条样式，透明度)
plt.grid(True,linestyle="--",alpha=0.5)

# 2.3 添加描述信息
plt.xlabel("时间",fontsize=20) #x轴标签
plt.ylabel("温度",fontsize=20) #y轴标签
plt.title("中午11点~12点某城市温度变化图",fontsize=20) #图标标签

# 2.5 显示图例 legend:图例
plt.legend(loc=0) #loc=0~10,对应图例显示的位置;也可以是“best”,"upper right","upper left","center"等

# 3.显示图像
plt.show() #画图并释放资源
```
